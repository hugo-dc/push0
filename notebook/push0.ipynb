```python
import os
import pandas as pd
```

```python
INPUT_FILE = '../data/5_histogram.csv'
```

```python
def read_file(fpath):
    fobj = open(fpath, 'r')
    data = fobj.read()
    fobj.close()
    return [line.split(',') for line in data.split('\n')]
```

```python
data = read_file(INPUT_FILE)
```

```python
columns = data[0]
```

```python
df = pd.DataFrame(data[2:-1], columns=columns)
```

```python
df['ALLPUSH'] = df['PUSH1'] + df['PUSH2'] +  df['PUSH3'] + df['PUSH4']  \
    + df['PUSH5'] + df['PUSH6'] + df['PUSH7'] + df['PUSH8'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2'] \
    + df['PUSH1'] + df['PUSH2'] + df['PUSH1'] + df['PUSH2']
```
